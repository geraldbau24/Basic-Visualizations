```r
if("ggplot2" %in% rownames(installed.packages()) == FALSE) {install.packages("ggplot2")}
library(ggplot2)
```

```r
qplot(mpg, wt, data=mtcars)
```

```r
ggplot(mtcars, aes(x = mpg, y = wt)) + geom_point(shape=1)
```

```r
ggplot(mtcars,aes(x=mpg,y=wt)) + geom_point(shape=19)
```

```r
ggplot(mtcars,aes(x=mpg,y=wt,shape = cyl)) + geom_point() + scale_shape_identity()
```

```r
mtcars$cylFactor<- factor(mtcars$cyl)
```

```r
ggplot(mtcars,aes(x=mpg,y=wt,shape = cylFactor)) + geom_point()
```

```r
ggplot(mtcars,aes(x=mpg,y=wt)) + geom_point(shape=19, colour="blue")
```

```r
ggplot(mtcars,aes(x=mpg,y=wt,color = cyl))+ geom_point(shape=19)
```

```r
ggplot(mtcars,aes(x=mpg,y=wt,color = cylFactor)) + geom_point(shape=19)
```

```r
ggplot(mtcars,aes(x=mpg,y=wt,color = cylFactor)) + geom_point(shape=19) + labs(colour = "Cylinders")
```

```r
ggplot(mtcars,aes(x=mpg,y=wt,color = cylFactor)) + geom_point(shape=19) + xlab("Miles per Gallon ") + ylab("Weight") + labs(colour = "Cylinders") + ggtitle("Scatterplot")
```

```r
EuStockDF <- as.data.frame(EuStockMarkets)
```

```r
head(EuStockDF)
```

```r
ggplot(EuStockDF,aes(x=c(1:nrow(EuStockDF)), y = DAX)) + geom_line()
```

```r
ggplot(EuStockDF,aes(x=c(1:nrow(EuStockDF)), y = DAX)) + geom_line(size=1.5) + labs(x = "Stocks")
```

```r
ggplot(EuStockDF,aes(x=c(1:nrow(EuStockDF)), y = DAX)) + geom_line(size=1.5, colour="light blue") + labs(x = "Time", y = "Stocks")
```

```r
dax_smi_plot <- ggplot() + geom_line(data = EuStockDF,aes(x=c(1:nrow(EuStockDF)), y = DAX), size = 1.5, colour="light blue") + geom_line(data = EuStockDF,aes(x=c(1:nrow(EuStockDF)), y = SMI), size = 1.5, colour = "red") + labs(x = "Time", y = "Stocks")
print(dax_smi_plot)
```

```r
all_stocks <- ggplot() + geom_line(data = EuStockDF,aes(x=c(1:nrow(EuStockDF)), y = DAX), size=1, colour="light blue") + geom_line(data = EuStockDF,aes(x=c(1:nrow(EuStockDF)), y = SMI), size =1, colour = "red") +  geom_line(data = EuStockDF,aes(x=c(1:nrow(EuStockDF)), y = CAC), size =1, colour = "purple") +  geom_line(data = EuStockDF,aes(x=c(1:nrow(EuStockDF)), y = FTSE), size =1, colour = "green") + labs(x = "Time", y = "Stocks")
print(all_stocks)
```

```r
legend_stocks <- all_stocks + xlab("Days") + ylab("Price") + ggtitle("Eu Stocks")
print(legend_stocks)
```

```r
ggplot(mtcars,aes(x=mpg,y=wt)) + geom_point(shape=19) + geom_smooth(method="lm", se= FALSE, color = "red")
```

```r
ggplot(mtcars,aes(x=mpg,y=wt,color = cylFactor)) + geom_point(shape=19) + geom_smooth(method="lm", se= TRUE, color = "red")
```

```r
ggplot(mtcars,aes(x=mpg,y=wt,color = cylFactor)) + geom_point(shape=19) + geom_smooth(method="lm", se= TRUE, color = "red") + xlab("Miles per Gallon ") + ylab("Weight") + labs(colour = "Cylinders") + ggtitle("Linear Regression")
```

```r
ggplot(mtcars,aes(x=mpg,y=wt,color = cylFactor)) + geom_point(shape=19) + geom_smooth(method="auto", se= TRUE, color = "red") + xlab("Miles per Gallon ") + ylab("Weight") + labs(colour = "Cylinders") + ggtitle("Gaussian Regression")
```
